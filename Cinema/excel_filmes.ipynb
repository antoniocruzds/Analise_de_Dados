  ### ANÁLISE DE FILMES EXIBIDOS NO BRASIL DE 2009 A 2018 ###
Com base nos dados da Ancine (https://oca.ancine.gov.br/cinema) analisei os filmes nacionais e internacionais exibidos no Brasil entre 2009 e 2018;

arquivo -> Filmes_2009_2018.xlsx

```python
# Importando as bibliotecas
import pandas as pd
# Configurando o matplotlib #pedindo pra mostrar o gráfico no proprio notbook
%matplotlib inline 
```

```python
# Criando o DataFrame com base em um arquivo Exel
# pedindo pra ele pular as 2 primeiras linhas (skiprows)
df = pd.read_excel('2120.xlsx', skiprows=2)

# Exibindo as 5 primeiras linhas
df.head(5)
```

```python
# Exibir as ultimas 20 linhas
df.tail(20)
```

Podemos notar que a tabela tem uma deformidade nas ultimas linhas, e precisamos remover;

```python
#Recriando o dataframe sem as 2 primeiras e as 16 ultimas linhaas
df = pd.read_excel('2120.xlsx', skiprows=2, skipfooter=16)

#Pedindo pra ele exibir as 20 ultimas
df.tail(20)


```

```python
# Verificar todos os tipos das colunas
df.dtypes
```

Analisando a saida, podemos perceber que Data de lançamento está como obj e deveria ser um obj numerico 

```python
#Convertendo o Campo Data de lançamento
df['Data de lançamento'] = pd.to_datetime(df['Data de lançamento'])
```

Deu erro pois dentro de Data de lançamento existe alguma informação que n é número mas sim um srt (relançamento) que não previmos antes

```python
#Identificamos o problema na coluna Data de lançamento
df['Data de lançamento'].value_counts()
```

```python
# Exibir alguns filmes que tenham Relançamento na coluna de Data de lançamento
#lendo as 5 primeiras, pois o doc tem 175 variaveis com esse str
df[df['Data de lançamento'] == 'Relançamento'].head()
```

# Decisão sobre dados incorretos/faltantes/divergentes

Agora precisamos saber  que fazer com esses tipos de dados (descartar, arrumar, etc)

```python
# verificando o tamanho do arquivo para saber a relação entre os dados existentes, e os com 'errp'
df.shape
```

De acordo com o tamanho encontrado, percebemos que a relação é de que entre 6398 linhas de codigo, 175 estão apresentando erros.

Partindo dessa relação, decidi descartar esses valores

```python
# Criando um novo df sem os relançamentos # != (diferente)
df_novo = df[df['Data de lançamento'] != 'Relançamento'].copy()

# Verificar todos os tipos das colunas
df_novo.dtypes
```

Apareceu que a coluna é objeto, mas tecnicamente os compos com str 'relançamento' já foram excluidos

```python
#Convertendo o Campo Data de lançamento para datetime
df_novo['Data de lançamento'] = pd.to_datetime(df_novo['Data de lançamento'])
```

```python
# Verificar todos os tipos das colunas e se a coluna foi convertida
df_novo.dtypes
```

Agora vamos analisar essa coluna em especifico e verificar o ano que mais teve lançamentos

```python
#Qual ano mais teve lançamento?
df_novo['Data de lançamento'].dt.year.value_counts()
#selecionando apenas o ano e dps contando
```

Agora vamos plotar um grafico com essas informações

```python
# Criando um grafico
df_novo['Data de lançamento'].dt.year.value_counts().plot.bar(title = 'Maior Quantidade de Lançamentos por Ano')
```

```python
#Verificando o filme de maior bilheretia
df_novo[df_novo['Renda (R$) no ano de exibição'] == df_novo['Renda (R$) no ano de exibição'].max()]
```

```python
#Também podemos verificar o top 5 maiores bilheterias do périodo da seguinte forma
df_novo.nlargest(5, 'Renda (R$) no ano de exibição')
```

```python
# Verificando o menos rentaveis
df_novo[df_novo['Renda (R$) no ano de exibição'] == df_novo['Renda (R$) no ano de exibição'].min()]
```

Podemos observar que nos foi retornado alguns dados em que a arrecadação foi nula. assim, concluimos que ou realmente o filme não teve bilheteria, ou a ancine não tem os determinados dados.

```python
#Verificando o top 30, pra tentar um valor != de 0
df_novo.nsmallest(30, 'Renda (R$) no ano de exibição')
```

```python
#Verificando a quantidade de filmes brasileiros e estrangeiros
df_novo['Nacionalidade da obra'].value_counts()
```

```python
#Plotando o grafico de relação entre as nacionalidades
df_novo['Nacionalidade da obra'].value_counts().plot.bar(title = 'Relação entre as produções Brasileiras X Estrangeiras')
```
