```python
# Quantidade de cursos por uf
df['uf_busca'].value_counts()
```

```python
#Qual o curso com a mensalidade mais alta?
df[df['mensalidade'] == df['mensalidade'].max()]
```

```python
#Qual o curso mais barato do brasil?
df[df['mensalidade'] == df['mensalidade'].min()]
```

```python
#Selecionando apenas algumas colunas p visualisar (uf, cidade, mensalidade)
df[df['mensalidade'] == df['mensalidade'].min()][['uf_busca', 'cidade_busca', 'mensalidade']]
```

```python
#Mostrar os cursos mais caros por UF
#mostrando o uf, cidade, faculdade, mensalidade e curso
df.loc[df.groupby('uf_busca')['mensalidade'].idxmax()][['uf_busca', 'cidade_busca', 'universidade_nome', 'nome', 'mensalidade']]
```

```python
#Selecionando mensalidades entre 500 e 1000
df[(df['mensalidade'] >500) & (df['mensalidade'] <1000)][['uf_busca', 'cidade_busca', 'universidade_nome', 'nome', 'mensalidade']]
```

```python
# Analisando as 5 maiores mensalidades (mostrando só algumas colunas)
df.nlargest(10, 'mensalidade')[['uf_busca', 'cidade_busca', 'universidade_nome', 'nome', 'mensalidade']]
```

```python
# Analisando as 10 menores mensalidades (mostrando só algumas colunas)
df.nsmallest(10, 'mensalidade')[['uf_busca', 'cidade_busca', 'universidade_nome', 'nome', 'mensalidade']]
```
